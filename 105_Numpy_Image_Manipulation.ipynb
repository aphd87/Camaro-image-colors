```python
import numpy as np
from skimage import io
import matplotlib.pyplot as plt
```

```python
camaro = io.imread("camaro.jpg")
```

```python
print(camaro) #Array with color intensities for each pixel
```

```python
camaro.shape #1200 rows, 1600 columns, 3D array with intensitiy values between 0-255
```

```python
plt.imshow(camaro)
plt.show()
```

```python
#Cropping the image
cropped = camaro[0:500,:,:] #: we want everything from each axis
plt.imshow(cropped)
plt.show()
```

```python
#So we cropped along y axis
```

```python
#Cropping the image
cropped = camaro[:,400:1000,:] #: we want everything from each axis
plt.imshow(cropped)
plt.show()
#Cropped along horizontal axis
```

```python
#Cropping the image
cropped = camaro[350:1100,200:1400,:] #: we want everything from each axis
plt.imshow(cropped)
plt.show()
```

```python
#Save to disk
io.imsave("camaro_cropped.jpg", cropped) #Saved this image 
```

```python
#Flip the image
#Vertical flip
vertical_flip = camaro[::-1,:,:] 
plt.imshow(vertical_flip)
plt.show()

io.imsave("camaro_vertical_flip.jpg", vertical_flip)

```

```python
#Flip the image
#Horizontal flip
horizontal_flip = camaro[:,::-1,:] 
plt.imshow(horizontal_flip)
plt.show()

io.imsave("camaro_horizontal_flip.jpg", horizontal_flip)
```

```python
#Color channels
#Red
red = np.zeros(camaro.shape, dtype = "uint8")
red[:,:,0] = camaro[:,:,0]
plt.imshow(red)
plt.show()
```

```python
#Green
green = np.zeros(camaro.shape, dtype = "uint8")
green[:,:,1] = camaro[:,:,1]
plt.imshow(green)
plt.show()
```

```python
#Blue

blue = np.zeros(camaro.shape, dtype = "uint8")
blue[:,:,2] = camaro[:,:,2]
plt.imshow(blue)
plt.show()
```

```python
#Rainbow
camaro_rainbow = np.vstack((red,green,blue))
plt.imshow(camaro_rainbow)
plt.show
```

```python
io.imsave("camaro_camaro_rainbow.jpg", camaro_rainbow)
```
